**Создайте новый признак `Age_category`, где категории 1,2,3 будут присвоены людям до 30, от 30 до 55 и от 55 лет соответственно.**

```python
df['Age_category'] = [1 if age<30 else 2 if age<55 else 3 for age in df.Age]
```

```python
df.drop('Age_category', axis = 1, inplace = True)
```

```python
df['Age_category'] = df['Age'].apply(lambda x: 1 if x<30 else 2 if x<55 else 3)
```

**С помощью `str` можно применять строковые методы к строковым столбцам.**

Найдите самое популярное имя среди пассажиров Титаника мужского пола 

```python
'mccarthy, mr. timothy j'.replace(',','')
```

```python
male_names = df.loc[df.Sex == 'male'].Name.str.lower()
for symb in [',','.','\'','"','(',')']:
    male_names = male_names.str.replace(symb, ' ')
male_names.str.split().explode().value_counts()
```
